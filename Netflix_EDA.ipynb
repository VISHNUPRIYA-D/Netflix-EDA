```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="darkgrid")
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df["date_added"] = df["date_added"].str.strip()
df["date_added"] = pd.to_datetime(df["date_added"])
df["date_added"]
```

```python
df=df.dropna(subset=["title"])
df
```

```python
df['director'] = df['director'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Not Listed')
df = df.drop_duplicates()
df.isnull().sum()
```

```python
df["year_added"]=df["date_added"].dt.year
df['month_added']=df["date_added"].dt.month
df[['title','date_added','year_added','month_added']].head()
```

```python
df['type'].value_counts()
```

```python
df['type'].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90,
    explode=[0.05,0],
    shadow=True
)
plt.title("Distribution of Movies and TV Shows on Netflix")
plt.ylabel('')
plt.show()
```

```python
df['year_added'].value_counts().sort_index().plot(
kind='bar',
figsize=(10,5),
title='Number of Netflix Titles Added per year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.show()
```

```python
df['listed_in'].value_counts().head(10).plot(
    kind='barh',
    figsize=(10,6),
    title="Top 10 most common genres on netflix"
)
plt.xlabel("number of Titles")
plt.ylabel("Genre")
plt.show()
```

```python
df['director'].value_counts().head(10).plot(
    kind='bar',
    title='Top 10 Directors on netflix',
    figsize=(10,5))
plt.xlabel('Directors')
plt.ylabel('Number of Titles')
plt.xticks(rotation=90)
plt.show()
```

```python
df['rating'].value_counts().plot(
    kind='bar',
    figsize=(10, 5),
    title='Distribution of Content Ratings'
)
plt.xlabel('Rating')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()
```

```python
df.groupby(['year_added', 'type']).size().unstack().plot(
    kind='line',
    figsize=(10, 6),
    title='Movies vs TV Shows Added Over Years'
)
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.xticks(rotation=40)
plt.grid(True)
plt.show()
```

```python
df['director'].value_counts().head(10).plot(
    kind='bar',
    figsize=(10, 5),
    title='Top 10 Directors on Netflix'
)
plt.xlabel('Director')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45,ha='center')
plt.show()
```

```python
df['rating'].value_counts().plot(
    kind='bar',
    figsize=(10, 5),
    title='Distribution of Content Ratings'
)
plt.xlabel('Rating')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()
```

```python

# 1. Extract digits (using raw string)
df['duration_int'] = df['duration'].str.extract(r'(\d+)')

# 2. Convert to numeric, invalid entries become NaN
df['duration_int'] = pd.to_numeric(df['duration_int'], errors='coerce')

# 3a. Remove rows where you canâ€™t parse a number
df = df.dropna(subset=['duration_int'])

# 3b. Convert column to integer
df['duration_int'] = df['duration_int'].astype(int)

# 4. Run the analysis
df.groupby('type')['duration_int'].describe()
```

```python
directors_count=df['director'].value_counts().head(10)
directors_count
```

```python
directors_count.plot(
    kind='bar',
    figsize=(12,5),
    title='Top 10 directors on netflix'
)
plt.xlabel('Directors')
plt.ylabel('Titles')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
df['first_country']=df['country'].str.split(",").str[0].str.strip()
df['first_country'].head()
```

```python
top_countries=df['first_country'].value_counts().head(10)
top_countries
```

```python
top_countries.plot(
    kind='barh',
    figsize=(12,5),
    title="Top 10 countries on netflix")
plt.xlabel('Number of Titles')
plt.ylabel('Countries')
plt.show()
```
